# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## Column 수정

```python
df['학교'].replace({'북산고':'상북고'}, inplace=True)
```

```python
df
```

```python
df['학교'].replace({'상북고':'북산고', '능남고':'무슨고'}, inplace=True)
```

```python
df
```

```python
df['학교'].replace({'북산고':'상북고', '무슨고':'능남고'}, inplace=True)
```

```python
df
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교'
```

```python
df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
```

```python
df
```

```python
df['결과'] = "Fail"
```

```python
df.loc[df['총합'] > 400, '결과'] = "Pass"
```

```python
df
```

## Column 삭제

```python
df.drop(columns = ['총합']) # 총합 column을 삭제
```

```python
df.drop(columns = ['국어', '영어', '수학']) # 국어 영어 수학 컬럼을 삭제
```
