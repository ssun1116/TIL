# 8. 데이터 선택 (iloc)
i : integer location. 정수 위치를 가지고 데이터 가져옴.

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.iloc[0:4] ## 0행부터 3행까지의 데이터를 가져옴
```

```python
df.iloc[3:5, 2:4] # 마지막 끝부분을 포함하지 않음. 그 직전까지 가져옴.
```

```python
df.iloc[:,2:6] # 모든 행 가져옴.
```

```python
df.iloc[[0,1],2] # 1, 2행 학생의 3열 데이터
```
